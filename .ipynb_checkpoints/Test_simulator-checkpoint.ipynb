```python
uv_path=os.path.join(".", "stix_data", "uv.sav")
phcorr_path=os.path.join(".", "stix_data", "phase_corr.sav")
ps_path=os.path.join(".", "stix_data", "pitch_slit.sav")

# Coordinates of the (u,v)-points
uv_coordinates = readsav(uv_path)
u = uv_coordinates['u']
v = uv_coordinates['v']

# Phase correction factors
phase_corr = readsav(phcorr_path)
phase_corr = phase_corr['phase_corr']

# Pitch and slit values
pitch_slit_data = readsav(ps_path)

pf = pitch_slit_data['pf'] # Pitch front grid
pr = pitch_slit_data['pr'] # Pitch rear grid
sf = pitch_slit_data['sf'] # Slit front grid
sr = pitch_slit_data['sr'] # Slit rear grid

# Pixel area
pixel_area = 0.22*0.92
pixel_phase_factor = 45.

vis_arr = []
for i in range(1000):
    this_vis, sigamp = sm.ComputeVisFromCounts(xc, yc, FWHM_max, FWHM_min, flux, angle, u, v,
                                               sf, sr, pf, pr, phase_corr, pixel_area, pixel_phase_factor,
                                               add_noise=True)
    vis_arr.append(this_vis)
    
vis_no_noise, sigamp = sm.ComputeVisFromCounts(xc, yc, FWHM_max, FWHM_min, flux, angle, u, v,
                                               sf, sr, pf, pr, phase_corr, pixel_area, pixel_phase_factor,
                                               add_noise=False)
```

```python
for i in range(1000):
    plt.plot(np.sqrt(vis_arr[i][0:24]**2+vis_arr[i][24:48]**2),'g--')

# plt.plot(np.sqrt(bad_example[0:24]**2+bad_example[24:48]**2), 'b')
plt.plot(np.sqrt(vis_no_noise[0:24]**2+vis_no_noise[24:48]**2), 'y')
    
plt.show()
```
