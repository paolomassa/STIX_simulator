```python
import simulator as sm
from scipy.io import readsav
import os
import numpy as np
import matplotlib.pyplot as plt
```

# Simulator parameters

```python
uv_data = readsav(os.path.join('.','stix_data','uv.sav'))
u = uv_data['u']
v = uv_data['v']

phase_corr_data = readsav(os.path.join('.','stix_data','phase_corr.sav'))
phase_corr = phase_corr_data['phase_corr']

pitch_slit_data = readsav(os.path.join('.','stix_data','pitch_slit.sav'))
sf = pitch_slit_data['sf']
sr = pitch_slit_data['sr']
pf = pitch_slit_data['pf']
pr = pitch_slit_data['pr']

pixel_area = 0.22*0.92
pixel_phase_factor = 45.
```

```python
n_sources = 2
xc, yc, FWHM_max, FWHM_min, flux, orientation, vis, sigamp = sm.SimulateConfig(n_sources, u, v, sf, sr, pf, pr, phase_corr, pixel_area, pixel_phase_factor, add_noise = False)
```

# Create configuration of X-ray sources

```python
npix=128
pix=2.

x_ray_im = np.zeros((npix,npix))

for i in range(n_sources):
    x_ray_im += sm.CreateGaussianSource(xc[i], yc[i], FWHM_min[i], FWHM_max[i], flux[i], orientation[i], npix=npix, pix=pix)
```

```python
plt.imshow(x_ray_im)
plt.show()
```
